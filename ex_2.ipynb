```python
import pandas as ps
```

```python
data = ps.read_csv('https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=MSFT&apikey=8WAC2EQWT6CHVMKY&datatype=csv')
```

```python
data.timestamp = ps.to_datetime(data.timestamp)
```

```python
data.to_csv("file.csv")
```

```python
data['week'] = data.timestamp.apply(lambda x: x.week)
```

```python
data.head()
```

```python
data.groupby('week')[['timestamp','open']].max()
```

```python
step_3  = data.groupby('week')[['timestamp','open']].max()
```

```python
step_3.to_csv('step_3.csv')
```

```python
data.head()
```

```python
data['7d_mavg'] = data.iloc[:,1:2].rolling(window=7).mean()
```

```python
step_4 = data[data['close'] > data['7d_mavg']]
```

```python
step_4.to_csv('step_4.csv')
```
