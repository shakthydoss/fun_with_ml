```python
import pandas as ps 
```

```python
data = ps.read_csv("ticker_ctxs_v2_missing_days.csv")
```

```python
data['Date'] = ps.to_datetime(data.Date)
```

```python
data.head(n=15)
```

```python
data['dayofweek'] = data.Date.apply(lambda x: x.dayofweek)
```

```python
data.head(n=15)
```

```python
full_date_list = ps.date_range(start='2007-01-04', end='2016-12-06')
```

```python
df_full_date_list = ps.DataFrame({
    "Date" : full_date_list, 
    "dayofweek":full_date_list.dayofweek
})
```

```python
df_full_date_list.head(16)
```

```python
non_sun_and_sat_date =  df_full_date_list[(df_full_date_list['dayofweek'] != 5) & (df_full_date_list['dayofweek'] != 6)]
```

```python
non_sun_and_sat_date.count()
```
